# Project Summary

The primary objective of this project was to develop a machine learning model capable of accurately classifying breast cancer tumors into malignant or benign categories. This was achieved by utilizing the Breast Cancer Wisconsin (Diagnostic) dataset, which includes various measurements related to breast cancer cell nuclei characteristics.

## Data Preprocessing
The dataset was loaded into a pandas DataFrame, combining both features and the target variable (indicating whether the cancer is malignant or benign). The columns were named appropriately using the dataset's feature names, with an additional 'target' column.

```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.preprocessing import MinMaxScaler
import plotly.graph_objs as go
```

```python
# Load dataset
cancer = load_breast_cancer()

# Create DataFrame
df_cancer = pd.DataFrame(cancer['data'], columns=cancer['feature_names'])
df_cancer['target'] = cancer['target']
```

## Data Preparation for Modeling
The features and target variable were separated into X (features) and y (target) datasets. The 'target' column was dropped from X, and y was set to just the 'target' column.

```python
# Define X and y for model training
X = df_cancer.drop(['target'], axis=1)
y = df_cancer['target']

# Split the dataset into training and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

## Model Selection and Training
An SVC (Support Vector Classifier) model was trained on the scaled training data. This model is well-suited for classification tasks like this, where the goal is to distinguish between two categories.

```python
# Fit SVC model to the training set
svc_model = SVC()
svc_model.fit(X_train, y_train)

# Get predicted y from fitted model
y_predict = svc_model.predict(X_test)

# Use confusion matrix to evaluate the model performance
cm = confusion_matrix(y_test, y_predict)

# Use heatmap for better visualization
sns.heatmap(cm, annot=True)
plt.savefig("confusion matrix 1.png")
```

```python
cm
```

## Feature Scaling
Feature scaling was applied to both training and testing sets to normalize the data, ensuring that no variable would dominate the model due to its scale.

```python
# Scale features
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Convert scaled arrays back to DataFrame
X_train_scaled_df = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled_df = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```

```python
# Before & After Scaling
sns.scatterplot(x=X_train['mean area'],
               y=X_train['mean smoothness'],
               hue=y_train)
plt.savefig("mean area and smoothness before scaling.png")
```

```python
sns.scatterplot(x=X_train_scaled_df['mean area'],
               y=X_train_scaled_df['mean smoothness'],
               hue=y_train)
plt.savefig("mean area and smoothness after scaling.png")
```

```python
# Fit model on scaled dataset for improvement
svc_model.fit(X_train_scaled, y_train)
y_predict = svc_model.predict(X_test_scaled)
cm = confusion_matrix(y_test, y_predict)
sns.heatmap(cm, annot=True)
plt.savefig("confusion matrix 2.png")
```

```python
cm
```

```python
print(classification_report(y_test, y_predict))
```

## Parameter Tuning and Optimization
To improve model performance, I employed grid search with cross-validation (GridSearchCV) to find the optimal parameters for the SVC model. The parameters tuned included C (regularization parameter), gamma (kernel coefficient for 'rbf'), and the kernel type.

The best parameters were then used to train the SVC model, and the model's performance was evaluated again, showing improvement with 105 true positives, 3 false negatives, 0 false positive, and 63 true negatives in the confusion matrix.


```python
# Hyperparameter Tuning - find the best 'C' and 'gamma'
param_grid = {'C': [0.1, 1, 10, 100],
             'gamma': [1, 0.1, 0.01, 0.001],
             'kernel':['rbf']}

grid = GridSearchCV(SVC(), param_grid)
grid.fit(X_train_scaled, y_train)
```

```python
# Get the best performing parameter values
grid.best_params_
```

```python
# Predict the scaled test data
grid_predictions = grid.predict(X_test_scaled)

# Show confusion matrix for performance comparison
cm = confusion_matrix(y_test, grid_predictions)
sns.heatmap(cm, annot=True)
plt.savefig("confusion matrix 3.png")
```

```python
cm
```

```python
print(classification_report(y_test, grid_predictions))
```
